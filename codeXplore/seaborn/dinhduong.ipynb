Tạo 1 DataFrame mới theo các cột cần quan tâm

```python
'Methol 1: Remove các cột ko quan tâm';
col_remove= ['shelf', 'weight',	'cups',	'rating']
cereals_df_1= cereals.drop(col_remove, axis= 1)
cereals_df_1.head()
```

```python
'Methol 2: Tạo DataFram new';
new_col = ['name', 'mfr', 'type', 'calories', 'protein', 'fat',	'sodium', 'fiber',	'carbo',	'sugars', 'potass', 'vitamins']
cereals_df = pd.DataFrame(data= cereals, columns= new_col)
cereals_df.head()
```

```python
'Tìm hiểu về hàm corr tính ma trận cho bảng với số liệu là số (Numerals)';
# Tìm mối quan hệ giữa các dữ liệu và cột: Hàm corr nhân các ma trận hàm lượmh dinh dưỡng với nhau
```

```python
cereals_df_corr = cereals_df.corr()
cereals_df_corr.head()
```

Quan sát số liệu:
 + 2 số liệu giống nhau và đối xứng qua đường chéo.
 + Do đó, cần bỏ 1 phần của trên để dễ quan sát

```python
# Tạo ma trận onelike, có giá trị True
Matric_onces = np.ones_like(cereals_df_corr, dtype= bool)
Matric_onces
```

```python
Matric_onces.shape, cereals_df_corr.shape
```

```python
# Dùng hàm triu để thay đổi nữa trên là False
hafl_bool = np.triu(Matric_onces)
hafl_bool
```

Vẽ dữ liệu bằng Heatmap

```python
heat_map = sns.heatmap(cereals_df_corr, mask= hafl_bool) #mask: mặt nạ, che kín
heat_map
```

```python
# Tại Dòng 1, cột cuối dữ liệu ko có ý nghĩa nên cắt bỏ, mask[1:,:-1]
mask = hafl_bool[1:, :-1]
mask
```

```python
cereals_mask = cereals_df_corr.iloc[1:, :-1]
cereals_mask

#  Phóng to bảng đồ
fig, ax = plt.subplots(figsize=(8,6))

sns.set_theme() #Thay đổi nền xám
result_map = sns.heatmap(cereals_mask, mask= mask, cmap= 'Greens', annot=True)
# annot=True, ghi giá trị biểu đồ
```

```python
# là, tròn 
result_map =sns.heatmap(cereals_mask, mask= mask, cmap= 'Greens', annot=True, fmt='.2f')
```
